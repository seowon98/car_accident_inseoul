```python
import folium

folium.Map()
```

```python
df_seoul["위도"].mean()
df_seoul["경도"].mean()
```

```python
sl=[df_seoul["위도"].mean(),
df_seoul["경도"].mean()] #list
```

```python
map=folium.Map(sl,zoomstart=7)
map
```

```python
for n in df_seoul.index:
    casualty=df_seoul.loc[n,"사망자수"]
    injured=df_seoul.loc[n,"부상자수"]
    popup = f"casualty: {casualty}, injured: {injured}"
    
    location=[df_seoul.loc[n,"위도"],df_seoul.loc[n,"경도"]]
    folium.Marker(
        location=location,
        popup=popup,
    ).add_to(map)
map
```
